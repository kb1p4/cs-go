```python
import pandas as pd
```

```python
train_ds = pd.read_csv('train.csv')
train_ds.info()
train_ds.head(100)
```

```python
player_ds = pd.read_csv('players_feats.csv')
player_ds.info()
player_ds.head(100)
```

```python
len(player_ds['map_name'].unique())
```

```python
feats = ['id','total_kills','headshots','total_deaths','kd_ratio','damage_per_round','grenade_damage_per_round','maps_played','rounds_played','kills_per_round','assists_per_round','deaths_per_round','saved_by_teammate_per_round','saved_teammates_per_round','rating','kill_death','kill_round','rounds_with_kills','kill_death_difference','total_opening_kills','total_opening_deaths','opening_kill_ratio','opening_kill_rating','team_win_percent_after_first_kill','first_kill_in_won_rounds']
maps = ['Dust2','Vertigo','Nuke','Mirage','Overpass','Inferno','Ancient']
pl_lst = []
for i in range(len(player_ds)):
    for j in range(5):
        pl_tmp = []
        for k in range(len(feats)):
            pl_tmp.append(player_ds.iloc[i]['p'+str(j+1)+'_'+feats[k]])
        pl_tmp.append(player_ds.iloc[i]['team_id'])
        pl_tmp.append(maps.index(player_ds.iloc[i]['map_name'])+1)
        pl_lst.append(pl_tmp)
pldf = pd.DataFrame(data=pl_lst,columns=feats + ['team_id','map_id'])
pldf.head(100)
```

```python
pldf['headshots_per_round'] = pldf['headshots'] / pldf['rounds_played']
pldf['rounds_with_kills'] = pldf['rounds_with_kills'] / pldf['rounds_played']
pldf['opening_kills_per_round'] = pldf['total_opening_kills'] / pldf['rounds_played']
pldf['opening_deaths_per_round'] = pldf['total_opening_deaths'] / pldf['rounds_played']
pldf['first_kill_in_won_rounds_per_round'] = pldf['first_kill_in_won_rounds'] / pldf['rounds_played']
pldf.drop(['id','total_kills','headshots','total_deaths','kd_ratio','maps_played','rounds_played','kill_death_difference','rating'],axis = 1, inplace=True)
pldf.head(100)
```

```python
import numpy as np
pldf.insert(0,'team_map_id',np.int32)
for i in range(len(pldf)):
   pldf.loc[i,'team_map_id'] = pldf.loc[i,'team_id'] + 100000 * pldf.loc[i,'map_id']
pldf.drop(['team_id','map_id'],axis = 1, inplace=True)
pldf.head(100)
```

```python
team_map_stats = pldf.groupby(['team_map_id']).mean()
team_map_stats.head(100)
```

```python
print(team_map_stats.loc[104494])
team_map_stats.loc[104494][1]
```

```python
data_X = []
data_Y = []
for i in range(len(train_ds)-100):
    vector = []
    data_Y.append(train_ds.iloc[i]['who_win'])
    team1_map_id = train_ds.loc[i,'team1_id'] + 100000 * (maps.index(train_ds.loc[i,'map_name'])+1)
    team2_map_id = train_ds.loc[i,'team2_id'] + 100000 * (maps.index(train_ds.loc[i,'map_name'])+1)
    for j in range(20):
        vector.append(team_map_stats.loc[team1_map_id][j] - team_map_stats.loc[team2_map_id][j])
    data_X.append(vector)
print(data_X)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
```

```python
model = CatBoostRegressor()
model.fit(data_X, data_Y)
```

```python
test1 = pd.read_csv('test.csv')
test1.info()
test1.head(100)
```

```python
result1 = []
for i in range(len(test1)):
    vector = []
    team1_map_id = test1.loc[i,'team1_id'] + 100000 * (maps.index(test1.loc[i,'map_name'])+1)
    team2_map_id = test1.loc[i,'team2_id'] + 100000 * (maps.index(test1.loc[i,'map_name'])+1)
    for j in range(20):
        vector.append(team_map_stats.loc[team1_map_id][j] - team_map_stats.loc[team2_map_id][j])
    result1.append(round(model.predict(vector)))
test1.insert(5,'who_win',result1)
```

```python
print(test1)
test1.to_csv('test1.csv')
```

```python
to_del = []
for i in range(len(data_X)):
    nan_OK = True
    j = 0
    while (j<19) and nan_OK:
        nan_OK = not np.isnan(data_X[i][j])
        j+=1
    if not nan_OK:
        to_del.append(i)
for i in range(len(to_del)):
    data_X.pop(to_del[i] - i)
    data_Y.pop(to_del[i] - i)
```
